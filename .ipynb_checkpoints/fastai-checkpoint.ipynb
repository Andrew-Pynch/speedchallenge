```python
import cv2

from fastai import *
from fastai.vision import *
from fastai.metrics import *

from SpeedDataset import SpeedDataset
```

```python
fname = 'labels.csv'
path_img = 'vidCaps/'
```

```python
df = pd.read_csv(fname)
df.head()
```

```python
data = (ImageList.from_df(df, path_img, cols='image')
        .split_by_rand_pct()
        .label_from_df('label')
        .databunch()
       )
data.batch_size = 8
```

```python
data.show_batch(rows=3, figsize=(7,6))
```

```python
data
```

```python
learn = cnn_learner(data, models.resnet152, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-5))
```

```python
learn.save('stage-2')
```

```python
learn.export()
```
